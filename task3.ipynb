# 1.Maximum Product of Two Elements in an Array

```python
class Solution:
    def maxProduct(self, nums) :
        max_value = 0

        for i in range(len(nums)):
            for j in range(i + 1, len(nums)):
                current = (nums[i] - 1) * (nums[j] - 1)
                if current > max_value:
                    max_value = current

        return max_value

print(Solution().maxProduct([3,4,5,2]))
print(Solution().maxProduct([1,5,4,5]))
print(Solution().maxProduct([3,7]))
```

# 2.Count Number of Teams

```python
class Solution:
    def numTeams(self, rating):
             n = len(rating)
             count = 0

             for i in range(n):
                for j in range(i + 1, n):
                    for k in range(j + 1, n):
                        if (rating[i] < rating[j] < rating[k]) or (rating[i] > rating[j] > rating[k]):
                            count += 1

             return count



print(Solution().numTeams([2,5,3,4,1]))
print(Solution().numTeams([2,1,3]))
print(Solution().numTeams([1,2,3,4]))
```

# 3.Number of Students Doing Homework at a Given Time

```python
class Solution:
    def busyStudent(self, startTime, endTime, queryTime):
        count = 0

        for i in range(len(startTime)):
            if startTime[i] <= queryTime <= endTime[i]:
                count += 1

        return count



print(Solution().busyStudent([1,2,3], [3,2,7], 4))
print(Solution().busyStudent([4], [4], 4))
```

# 4.Number of Steps to Reduce a Number to Zero

```python
class Solution:
    def numberOfSteps(self, num: int) -> int:
        count = 0
        currnum = num

        while currnum > 0:
            if currnum % 2 == 0:   # even
                currnum //= 2
            else:                 # odd
                currnum -= 1
            count += 1

        return count


print(Solution().numberOfSteps(14))
print(Solution().numberOfSteps(8))
```

# 5.Counting Bits


```python
class Solution:
    def countBits(self, n: int):
        ans = []
        for i in range(n + 1):
            binary = bin(i)        # convert number to binary string
            ans.append(binary.count('1'))  # count '1's
        return ans


# Testing
print(Solution().countBits(2))  # [0,1,1]
print(Solution().countBits(5))  # [0,1,1,2,1,2]
```
